```python
class MyError(Exception):
    pass

def say_nick(nick):
    if nick == "돼지":
        raise MyError() #에러를 발생시킨다.
    print(nick)

try:
    say_nick("곰")
    say_nick("돼지")
    
except MyError:
    print('허용하지 않는 별명입니다.')
```

```python
try:
    print("파일 열기")
    f = open('foo.txt', 'z')
    # 동작할 코드 내용 작성    

finally:
    print('파일 닫기')
    f.close()    # try 블럭 수행 중간에 오류가 발생하더라도 마지막에 무조건 실행된다.
```

```python
try:
    age=int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
```
