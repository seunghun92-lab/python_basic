```python
from enum import Enum

```

```python
# name = value 지정 필수
class Gender(Enum):
    M = '남'
    F = '여'

print(Gender.M)
print(Gender.F)
```

```python
# name
print(Gender.M.name)
print(Gender.F.name)

# value
print(Gender.M.value)
print(Gender.F.value)
```

```python
a = Gender.M
b = Gender.M

print(a == b) #값 비교
print(a is b) #객체 비교 / 동일한 인스턴스인지 확인 
```

```python
print(Gender.M == 'M') # False
print(Gender.M.name == 'M') # True

a = Gender.M
b = Gender.M
print(a == b) # True 단일 인스턴스
```

```python
# iteration
for gender in Gender:
    print(gender)
```

```python
Color = Enum('Color', ['RED', 'GREEN', 'BLUE'])
for color in Color:
    print(color.name, color.value)
    
```

```python
from enum import auto

class Lang(Enum):
    # auto()로 지정되는 값을 override 가능
    # member선언보다 먼저 작성할 것
    def _generate_next_value_(name, start, count, last_value):
        # return name # name, value 동일
        return start # 첫번째 요소값 - 동일한 value가 지정되는 경우 선언무시된다?
        # return count # 0 1 2 인덱스 지정
        # return last_value # 마지막에 지정된 value

    C = auto()        # 1
    Java = auto()     # 2
    Python = auto()   # 3

for lang in Lang:
    print(lang.name, lang.value)
```

```python
# member.value가 동일한 경우, 선언이 무시된다?
class BookCategory(Enum):
    인문 = None
    소설 = None
    시 = None

for category in BookCategory:
    print(category.name, category.value)
```

```python
# IF / MATCH 분기를 안전하게 하고싶을때 사용!
#기존 방식
status = "RUNNIG"

if status == "RUNNING":
    pass
elif status == "READY":
    pass
else:
    pass

# else 분기문을 타게 된다.
```

```python
class Status(Enum):
    RUNNING = "RUNNNG"

status = Status.RUNNING

if status == status.RUNNING:
    pass
```
