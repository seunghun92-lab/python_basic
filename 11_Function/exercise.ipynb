2. 랜덤하게 하나의 메뉴를 선택하고 반환하세요.
    - 출력 예시: **"오늘의 점심 메뉴는 [선택된 메뉴]입니다!"**
    - 힌트: `random.choice()`를 사용하면 랜덤 선택이 가능합니다.

```python
import random

def recommend_menu(*menus):
    choice_menu = random.choice(menus)
    return choice_menu


result = recommend_menu('김밥', '라면', '구내식당', '돈까스')
print(f"오늘의 점심 메뉴는 [{result}]메뉴 입니다!")
```

**다람쥐는 오늘의 업무 성과를 보고해야 한다."**

1. `work_report(tasks_done)`이라는 함수를 만들어 오늘 처리한 업무 수를 매개변수로 받으세요.
2. 함수는 다음 두 값을 반환합니다.
    - `tasks_done`: 처리한 업무 수
    - `status`: 업무 성과 평가 (10개 이상: "우수", 5~9개: "보통", 4개 이하: "노력 필요")
3. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"오늘 처리한 업무는 [업무 수]개입니다. 성과 평가: [평가]"**

```python
def work_report(tasks_done): 
    
    states = " "

    if tasks_done > 10:
        states = "우수"
    elif tasks_done > 5:
        states = '보통'
    else: 
        states = "노력 필요"
    return states, tasks_done

result = work_report(7)

print(f"오늘 처리한 업무는 {result[1]}개입니다.: {result[0]}")
```

## **문제 7: 월급 계산 및 세금 공제**

**"다람쥐는 첫 월급을 계산하고 세금을 공제해야 한다."**

1. `calculate_salary(base_salary, bonus)`라는 함수를 만들어 기본 월급과 보너스를 매개변수로 받으세요.
2. 총 월급을 계산하고, 세금은 총 월급의 10%로 공제하세요.
3. 함수는 다음 두 값을 반환합니다.
    - `net_salary`: 실수령액 (총 월급 - 세금)
    - `tax`: 세금 공제액
4. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"세전 월급: [총 월급]원, 세금: [세금]원, 실수령액: [실수령액]원"**

```python
# 1. calculate 함수 만들거다(입력 2개 받을거다)
# 2. 이거 두개 합쳐서 총 월급
# 3. 세금은 총월급의 10%
# 4. 실수령액은 총월급 - 세금 

# 출력값은 총월급, 세금, 실수령액


#. 함수 호출하기 (입력값 2개  - 기본월급, 보너스) 해서 나온 값 저장


# 출력(중간에 총월급, 세금, 실수령액 3개 f스트링으로 넣으면 될것같아.)
```

```python
def calculate_salary(base_salary, bonus):
    total_salary = base_salary + bonus
    tax = total_salary * 0.1
    net_salary = total_salary - tax
    return net_salary, tax, total_salary

result = calculate_salary(5_000_000, 2_000_000)
print(result)

print(f"세전 월급{result[0]}원, 세금:{result[2]}원, 실수령액: {result[1]}원")
```
